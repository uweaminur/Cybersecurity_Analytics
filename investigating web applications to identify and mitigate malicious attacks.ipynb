## <u> Step-2: Meaning of each error code </u> 
### From above list, i get only 4 error code and from this error code, I find below explananion for each code,

* Error code 200 = The request was received and understood and is being processed.
* Error code 404 = The page you were trying to reach on a website couldn't be found on their server
* Error code 301 = Moved Permanently is used for permanent redirecting.
* Error code 401 = The client request has not been completed because it lacks valid authentication credentials for the requested resource.

<div style="text-align: justify"> 
Above explanation of the codes, I understand that only one code explained the user who trying to access to the server is not the real user and who do not know the credentials. If the user knows the credential, then there will not be generated such code more than thousands time. Again, if the user is real and sometimes mistakenly input the wrong credential (which is natural) the error code would generates more than thousands time which is suspicious.
So, I start investigate with this error code 401. Below is the command to collect all the information related to 401 codes. 
<br><br>
Note that rest of the error code 404 and 301 are hits by more than four thousand times but the error explanation shows that wrong web address or redirecting accordingly. About error code 200 is for successful action. So, i continue with error code 401.

```python
data_401 = data[data['sc-status']==401]

data_401
```

##  <u> Step-3: Convert time </u>

### Here we use conversion to get the total minute.

```python
data_401['total_time'] = pd.to_datetime(data_401['time'])
# data_401['total_minutes'] = data_401['total_time'].dt.minute
data_401['hit_time'] = ((data_401['total_time'].dt.hour * 100) + data_401['total_time'].dt.minute)
data_401
```
